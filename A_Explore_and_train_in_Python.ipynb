```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

credit = pd.read_csv('data/creditcard.csv', header=None, names=list(range(1, 31)) + ['Class'])
credit.head(5)
```

```python
credit.describe().T
```

```python
credit.describe().T.iloc[1:-2,1:].plot()
plt.show()
```

```python
Z = (credit-credit.mean())/credit.std()
Z.describe().T
```

```python
Z.describe().T.iloc[1:-1,1:].plot()
plt.show()
```

```python
fig, axs = plt.subplots(10,3, figsize=(16,20))
for i in range(30):
    axs[i//3,i%3].plot(Z.iloc[:,i])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

#X = credit.drop('Class', axis=1).values
X = Z.drop('Class', axis=1).values
X = np.clip(X,-5.,5.)

y = credit.Class.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

print(X.min(), X.max())
```

```python
np.savetxt("data/X_train.csv", X_train, delimiter=",")
np.savetxt("data/X_test.csv", X_test, delimiter=",")
np.savetxt("data/y_train.csv", y_train, delimiter=",")
np.savetxt("data/y_test.csv", y_test, delimiter=",")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_hat = clf.predict(X_test)
roc_auc_score(y_test, y_hat)
```

```python
acc = np.isclose(y_test, y_hat).mean()
acc
```

```python
np.savetxt("data/coeff.csv", clf.coef_.reshape((-1,1)), delimiter=",")
np.savetxt("data/intercept.csv", np.array([clf.intercept_]), delimiter=",")
```

```python
clf.coef_, clf.intercept_, clf.coef_.min(), clf.coef_.max()
```

```python
clf.coef_.shape, X.shape
```

```python
terms = X*clf.coef_
print(terms.min(), terms.max())
#
pd.DataFrame(terms, columns=credit.columns[:-1]).describe().T
```

```python
temps = terms.cumsum(axis=1)
print(temps.min(), temps.max())
#
pd.DataFrame(temps, columns=credit.columns[:-1]).describe().T
```
